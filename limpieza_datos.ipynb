```python
import pandas as pd
import csv
```

```python
deaths2017 = pd.read_csv('./deaths.csv')
deaths2018 = pd.read_csv('./2018_defuncions_edats-quinquennals.csv')
deaths2019 = pd.read_csv('./2019_defuncions_edats-quinquennals.csv')
deaths2020 = pd.read_csv('./2020_defuncions_edats-quinquennals.csv')
deaths2017.head()
```

```python
deaths2017['Age'].unique()
```

```python
deaths2018.head()
```

```python
deaths2018['Edats_quinquennals'] = deaths2018['Edats_quinquennals'].str.replace('anys', '').str.replace('100','>=100').str.replace('i més','').str.strip()
deaths2019['Edats_quinquennals'] = deaths2019['Edats_quinquennals'].str.replace('anys', '').str.replace('100','>=100').str.replace('i més','').str.strip()
deaths2020['Edats_quinquennals'] = deaths2020['Edats_quinquennals'].str.replace('anys', '').str.replace('100','>=100').str.replace('i més','').str.strip()
```

```python
deaths2018['Edats_quinquennals'].unique()
```

```python
deaths2018 = deaths2018.rename({'Any': 'Year'
                                ,'Codi_districte': 'District.Code'
                                ,'Nom_districte': 'District.Name'
                                ,'Codi_barri': 'Neighborhood.Code'
                                ,'Nom_barri': 'Neighborhood.Name'
                                ,'Edats_quinquennals': 'Age'
                                ,'Nombre': 'Number'}, axis=1)  # new method
deaths2019 = deaths2019.rename({'Any': 'Year'
                                ,'Codi_Districte': 'District.Code'
                                ,'Nom_Districte': 'District.Name'
                                ,'Codi_Barri': 'Neighborhood.Code'
                                ,'Nom_Barri': 'Neighborhood.Name'
                                ,'Edats_quinquennals': 'Age'
                                ,'Nombre': 'Number'}, axis=1)  # new method
deaths2020 = deaths2020.rename({'Any': 'Year'
                                ,'Codi_Districte': 'District.Code'
                                ,'Nom_Districte': 'District.Name'
                                ,'Codi_Barri': 'Neighborhood.Code'
                                ,'Nom_Barri': 'Neighborhood.Name'
                                ,'Edats_quinquennals': 'Age'
                                ,'Nombre': 'Number'}, axis=1)  # new method
```

```python
deaths2018.head()
```

```python
deaths2019.head()
```

```python
deaths2020.head()
```

```python
vertical_stack = pd.concat([deaths2017, deaths2018], axis=0)
vertical_stack = pd.concat([vertical_stack, deaths2019], axis=0)
vertical_stack = pd.concat([vertical_stack, deaths2020], axis=0)
```

```python
vertical_stack['Year'].unique()
```

```python
vertical_stack.head()
```

```python
vertical_stack.to_csv('deaths_to_2020.csv', index=False,doublequote=True,quoting=csv.QUOTE_NONNUMERIC)
```
